```python
import pandas as pd
import ast
```

# Import dataframes 

```python
train = pd.read_csv('Data/train.csv')
#train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/kaggle/NBME_patient_notes2/Data/train.csv')
train['annotation'] = train['annotation'].apply(ast.literal_eval)
train['location'] = train['location'].apply(ast.literal_eval)
#features = pd.read_csv('../input/nbme-score-clinical-patient-notes/features.csv')
features = pd.read_csv('Data/features.csv')
#features = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/kaggle/NBME_patient_notes2/Data/features.csv')
def preprocess_features(features):
    features.loc[27, 'feature_text'] = "Last-Pap-smear-1-year-ago"
    return features
features = preprocess_features(features)
#patient_notes = pd.read_csv('../input/nbme-score-clinical-patient-notes/patient_notes.csv')
patient_notes = pd.read_csv('Data/patient_notes.csv')
display(train.head())
display(features.head())
display(patient_notes.head())
display(len(patient_notes))
```

```python
#delete from patient notes notes that exists in train
patient_notes= patient_notes.merge(train, on=['pn_num', 'case_num'], how='left')
patient_notes = patient_notes.drop(patient_notes[patient_notes['id'].isnull() == False].index)
patient_notes = patient_notes.drop(['id', 'feature_num','annotation','location'], axis=1)
#patient_notes = patient_notes.reset_index()
unlabeled = pd.DataFrame(columns = ['pn_num','case_num','pn_history'])
for case in range(10):
    new_data = patient_notes[patient_notes['case_num'] == case]
    new_data = new_data.sample(n = 60)
    unlabeled = pd.concat([unlabeled,new_data],ignore_index = True)
display(len(unlabeled))
```

```python
unlabeled = unlabeled.merge(features, on=['case_num'])
display(len(unlabeled))
```
